## Grfico de Barras

Representar dados categricos. Comparar quantidade ou frequncia de diferentes categorias monstrar diferenas em grupos e visualizar rapidamente quais categorias tm valores maiores ou menores. 

```python
x_2 = np.arange(1, 10, 2)
y_2 = [7, 8, 2, 4, 2]
plt.bar(x_2, y_2, label='Lista 2', color='red')
plt.legend()
plt.show()
```

## Grfico de Disperso
Relao entre duas variveis contnuas. Cada ponto representa um par de valores das duas variveis.

```python
x = np.arange(1, 9)
y = [5, 2, 4, 5, 6, 8, 4, 8]

plt.scatter(x, y, label='Pontos', color='black', marker='*')
plt.legend()
plt.show()
```

## Grfico de rea empilhada
Visualiza a mudana relativa de diversas variveis ao longo do tempo. Mostrar como as partes contribuem para o todo ao longo do tempo.

```python
dias = np.arange(1, 6)
dormir = [x for x in range(8, 5, -1)] + [77, 7]
comer = [y for y in range(2, 6)] + [3]
trabalhar = [7, 8, 7, 2, 2]
passear = [8, 5, 7, 8, 13]

plt.stackplot(dias, dormir, comer, trabalhar, passear, colors=['m', 'c', 'r', 'k', 'b'])
plt.show()
```

## Grfico de Pizza
Cada fatia do grfico corresponde a uma categoria e sua rea  proporcional  porcentagem que a categoria representa do todo. Mostrar a distribuio de dados em relao ao todo e para destacar as propores relativas de diferentes categorias.

```python
fatias = [7, 2, 2, 13]
atividades = ['dormir', 'comer', 'passear', 'trabalhar']
cores = ['olive', 'lime', 'violet', 'royalblue']

plt.pie(fatias, labels=atividades, colors=cores, startangle=90, shadow=True, explode=(0, .2, 0, 0))
```
