## Gráfico de Barras

Representar dados categóricos. Comparar quantidade ou frequência de diferentes categorias monstrar diferenças em grupos e visualizar rapidamente quais categorias têm valores maiores ou menores. 

```python
x_2 = np.arange(1, 10, 2)
y_2 = [7, 8, 2, 4, 2]
plt.bar(x_2, y_2, label='Lista 2', color='red')
plt.legend()
plt.show()
```

## Gráfico de Dispersão
Relação entre duas variáveis contínuas. Cada ponto representa um par de valores das duas variáveis.

```python
x = np.arange(1, 9)
y = [5, 2, 4, 5, 6, 8, 4, 8]

plt.scatter(x, y, label='Pontos', color='black', marker='*')
plt.legend()
plt.show()
```

## Gráfico de área empilhada
Visualiza a mudança relativa de diversas variáveis ao longo do tempo. Mostrar como as partes contribuem para o todo ao longo do tempo.

```python
dias = np.arange(1, 6)
dormir = [x for x in range(8, 5, -1)] + [77, 7]
comer = [y for y in range(2, 6)] + [3]
trabalhar = [7, 8, 7, 2, 2]
passear = [8, 5, 7, 8, 13]

plt.stackplot(dias, dormir, comer, trabalhar, passear, colors=['m', 'c', 'r', 'k', 'b'])
plt.show()
```

## Gráfico de Pizza
Cada fatia do gráfico corresponde a uma categoria e sua área é proporcional à porcentagem que a categoria representa do todo. Mostrar a distribuição de dados em relação ao todo e para destacar as proporções relativas de diferentes categorias.

```python
fatias = [7, 2, 2, 13]
atividades = ['dormir', 'comer', 'passear', 'trabalhar']
cores = ['olive', 'lime', 'violet', 'royalblue']

plt.pie(fatias, labels=atividades, colors=cores, startangle=90, shadow=True, explode=(0, .2, 0, 0))
```
